```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd

# Define the Lasso model
lasso = Lasso(alpha=0.1)  # You can tune alpha if needed

# Select predictors (exclude non-feature columns)
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]

# Backtest function (same as before)
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        # Handle missing values in train and test sets
        train = train.dropna()  # Drop rows with NaNs in training data
        test = test.dropna()  # Drop rows with NaNs in test data

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)

# Run backtest with Lasso
predictions_lasso = backtest(weather, lasso, predictors)

# Evaluate performance (handle potential NaNs in predictions)
mae = mean_absolute_error(predictions_lasso["actual"].dropna(), predictions_lasso["prediction"].dropna()) # Drop NaNs before calculating MAE
mse = mean_squared_error(predictions_lasso["actual"].dropna(), predictions_lasso["prediction"].dropna()) # Drop NaNs before calculating MSE

print(f"✅ Lasso MAE: {mae:.4f}")
print(f"📉 Lasso MSE: {mse:.4f}")
```

```python
def pct_diff(old, new): #calculate the percentage
    return (new - old) / old

def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col]) #check the percentage
    return weather

rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("updated_weather.csv")

# Convert day_avg_tmax and day_avg_tmin from Celsius to Fahrenheit
df['day_avg_tmax_f'] = (df['day_avg_tmax'] * 9/5) + 32
df['day_avg_tmin_f'] = (df['day_avg_tmin'] * 9/5) + 32

# Calculate the average temperature in Fahrenheit
df['tavg_f'] = (df['day_avg_tmax_f'] + df['day_avg_tmin_f']) / 2

# Calculate the saturation vapor pressure (e_s) in Fahrenheit
df['e_s_f'] = 6.11 * (10 ** (7.5 * df['tavg_f'] / (df['tavg_f'] + 237.3)))

# Estimate the actual vapor pressure (e)
df['e_f'] = df['e_s_f'] * (df['day_avg_tmin_f'] / df['day_avg_tmax_f'])

# Estimate humidity in percentage
df['humidity_estimate_percent_f'] = (df['e_f'] / df['e_s_f']) * 100

# Print the first few rows of the relevant columns
print(df[['day_avg_tmax_f', 'day_avg_tmin_f', 'humidity_estimate_percent_f']].head())

# Save the result to a new CSV
df.to_csv("humidity", index=False)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the data
df = pd.read_csv("humidity")
df = df.dropna()

# Features for prediction (temperature and humidity)
X = df[['day_avg_tmin', 'day_avg_tmax', 'humidity_estimate_percent_f']]

# Target for chance of rain prediction based on heuristic
df['will_rain'] = ((df['day_avg_tmax'] > 25) & (df['humidity_estimate_percent_f'] > 60)).astype(int)

# Train-test split for classification (chance of rainfall prediction)
X_train, X_test, y_train_rain_chance, y_test_rain_chance = train_test_split(X, df['will_rain'], test_size=0.2, random_state=42)

# Using LogisticRegression or another model for predicting rain chance
rain_chance_model = LinearRegression()  # For simplicity, using linear regression as a classifier
rain_chance_model.fit(X_train, y_train_rain_chance)

# 🔍 Input your own values
user_tmin = float(input("Enter day_avg_tmin: "))
user_tmax = float(input("Enter day_avg_tmax: "))
user_humidity = float(input("Enter humidity_estimate_percent_f: "))  # Use the correct column name here

# Predict chance of rainfall using the model
input_data = pd.DataFrame([[user_tmin, user_tmax, user_humidity]],
                          columns=['day_avg_tmin', 'day_avg_tmax', 'humidity_estimate_percent_f'])  # Ensure correct column name

rain_chance = rain_chance_model.predict(input_data)[0]  # Predicting the chance of rain

# Display result
print(f"\n☔ Chance of rainfall: {rain_chance:.2f}%")
```

```python
import pandas as pd

# Assuming you have a DataFrame named df with Fahrenheit values
def estimate_drought_flood_risk(row):
    tmax = row['day_avg_tmax_f']  # Using Fahrenheit temperatures
    tmin = row['day_avg_tmin_f']  # Using Fahrenheit temperatures
    humidity = row['humidity_estimate_percent_f']  # Using the estimated humidity in percentage
    e_s = row['e_s_f']  # Saturation vapor pressure in Fahrenheit
    e = row['e_f']  # Actual vapor pressure in Fahrenheit

    # Adjusting thresholds based on Fahrenheit logic
    # Example logic (you can tune the thresholds based on domain knowledge)
    drought_risk = (tmax > 104) and (humidity < 30) and (e_s > e)  # tmax > 104°F for drought
    flood_risk = (tmin < 77) and (humidity > 80) and (e > e_s)  # tmin < 77°F for flood

    return pd.Series({
        'drought_likely': drought_risk,
        'flood_likely': flood_risk
    })

# Apply to entire DataFrame with updated Fahrenheit values
df[['drought_likely', 'flood_likely']] = df.apply(estimate_drought_flood_risk, axis=1)

# Print the relevant columns
print(df[['day_avg_tmax_f', 'day_avg_tmin_f', 'humidity_estimate_percent_f', 'drought_likely', 'flood_likely']].head())
```
